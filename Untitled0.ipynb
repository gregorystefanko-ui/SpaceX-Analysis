```python
import requests
import pandas as pd
from google.colab import files
import time

all_launches = []
url = "https://lldev.thespacedevs.com/2.2.0/launch/previous/"

print("Fetching historical SpaceX launches...")

for offset in [0, 100, 200, 300]:
    params = {
        "lsp__name__icontains": "SpaceX",
        "limit": 100,
        "offset": offset
    }

    response = requests.get(url, params=params)
    if response.status_code == 200:
        data = response.json()
        results = data.get('results', [])
        all_launches.extend(results)
        print(f"Collected {len(all_launches)} launches...")
        if len(results) < 100:
            break
    else:
        print(f"Error: {response.status_code}")
        break
    time.sleep(1)

# Create the Excel file
df = pd.json_normalize(all_launches)
filename = "SpaceX_Clean_History.xlsx"
df.to_excel(filename, index=False)

print(f"Success. Total Rows: {len(df)}")
files.download(filename)
```
